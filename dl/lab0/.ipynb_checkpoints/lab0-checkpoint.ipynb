```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pmc_1.csv')
df
```

```python
df.head()
```

```python
df.describe()
```

```python
print("Missing values in the dataset:")
df = df.drop(columns=['Date'])
print(df.isnull().sum())
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
selected_attributes = ['dailysamples', 'dailyconfirmed', 'dailyrecovered', 'dailydeceased', 'totalcritical']

print("Sum of selected attributes:")
print(df[selected_attributes].sum())

print("Mean of selected attributes:")
print(df[selected_attributes].mean())

print("Median of selected attributes:")
print(df[selected_attributes].median())

print("Standard Deviation of selected attributes:")
print(df[selected_attributes].std())
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(df.index, df['totalconfirmed'], label='Total Confirmed', color='blue')
ax.bar(df.index, df['totalrecovered'], bottom=df['totalconfirmed'], label='Total Recovered', color='green')
ax.set_title('Total Confirmed vs Total Recovered')
ax.legend(labels=['Total Confirmed', 'Total Recovered'])
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```
